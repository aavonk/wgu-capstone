```python
import matplotlib as mpl
import pandas as pd
import re

movies = pd.read_csv("./data/movies.csv")

# Clean up the titles of each movies and keep only alpha numeric characters
def clean_title(title):
   return re.sub("[^a-zA-Z0-9 ]","", title)

def get_top_genre(genre):
    return genre.split("|")[0]


def get_production_year(title):
    last_digits =  title.strip()[-5:-1]
    # if there isn't a year, return a year
    if last_digits.isnumeric():
        return last_digits
    # If there is no year in the title, default to 1960
    else:
        return "1960"


movies["clean_title"] = movies["title"].apply(clean_title)
movies["production_year"] = movies["title"].apply(get_production_year)
movies["top_genre"] = movies["genres"].apply(get_top_genre)
```

## Most Popular Ratings
### The below histogram shows how many times a particular rating score was given across all movies

##### By looking at this data, we can see how often particular ratings are given, and which ones are more popular.


```python
ratings = pd.read_csv('./data/ratings.csv')
hist = ratings['rating'].hist(bins=20)
hist.set_xlabel("Rating Score", labelpad=20, weight='bold', size=12)
hist.set_ylabel("Frequency", labelpad=20, weight='bold', size=12)
```

```python
# Display all of the movies grouped by genre
unique_genres = movies["top_genre"].unique()[:8]

stats = pd.DataFrame([], columns=["genre", "count"])

for genre in unique_genres:
    count = 0;
    #count how many times this genre has appeared
    count = movies['top_genre'].value_counts()[genre]

    #add the item to the dataframe
    stats.loc[len(stats)] = [genre, count]
stats.set_index("genre", inplace=True)
stats.plot.pie(y="count", autopct='%1.1f%%', title="Top 8 Movie Genres", legend=False, shadow=True)
```

```python

condition = movies["production_year"].astype(int) < 2023

prodcount = (movies[condition][["production_year", "movieId"]]
             .groupby("production_year")
             .count()
            )

prodcount.tail(20)

(prodcount
 .rolling(5).mean().rename(columns={"movie_id":"count"})
 .plot(figsize=(15,5),
       title="Count of Rated Movies - by production year")
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2))
# use the vectorizer to turn our set of movie titles into a matrix of sets of numbers
tfidf = vectorizer.fit_transform(movies["clean_title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

#Find related movies based on the similarity of the title
def search_movie(title):
    title = clean_title(title)
    query_vec = vectorizer.transform([title])
    similarity_scores = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity_scores, -5)[-5:]
    results = movies.iloc[indices][::-1]
    return results
```

```python


# When a user searches a movie, we are going to search the data for all of the users who rated (and liked)
# the movie entered then, we will find other movies they like to recommend to the user.

def find_recommended_movies(movie_id):
    #Find users that have both watched the movie and given it a rating of 4 or more and other movies the recommend
    similar_user_ids = ratings[(ratings["movieId"] == movie_id) & (ratings["rating"] >= 4)]["userId"].unique()
    similar_user_recommendations = ratings[(ratings["userId"].isin(similar_user_ids) & (ratings["rating"] >=4 ))]["movieId"]

    # Filter the movies down to movies where 20% or more of users similar to us also liked
    similar_user_recommendations = similar_user_recommendations.value_counts() / len(similar_user_ids)
    similar_user_recommendations = similar_user_recommendations[similar_user_recommendations > .2]
    
    
    #Find all of the users who have watched the movie we searched for and what they recommend
    all_users = ratings[(ratings["movieId"].isin(similar_user_recommendations.index) & (ratings["rating"] >=4 ))]
    all_users_recs = all_users["movieId"].value_counts() / len(all_users["userId"].unique())
    
    
    #build a recommendation score and sort it
    
    rec_percentages = pd.concat([similar_user_recommendations, all_users_recs], axis=1)
    rec_percentages.columns = ["similar", "all"]

    rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
    rec_percentages = rec_percentages.sort_values("score", ascending=False)
    
    # Merge with our movie data and return
    return rec_percentages.head(10).merge(movies, left_index=True, right_on="movieId")

```

```python
import ipywidgets as widgets
from IPython.display import display

movie_input_name = widgets.Text(
    value="",
    description="Movie title",
    disabled=False
)

recs = widgets.Output()


def on_typing(data):
    with recs:
        recs.clear_output()
        title = data["new"]
        if len(title) >5:
            results = search_movie(title)
            movie_id = results.iloc[0]["movieId"]
            found_recommendations = find_recommended_movies(movie_id)

            display(found_recommendations[["score", "title", "genres"]])
            
           
            
movie_input_name.observe(on_typing, names="value")

```
